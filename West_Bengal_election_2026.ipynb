```python
#This bar chart shows the top 10 political parties based on the total number of votes received. It helps identify the most popular and influential parties in the election.
df.groupby('Party')['Total Votes'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(df.groupby('Party')['Total Votes'].sum().sort_values(ascending=False).head(10).index, df.groupby('Party')['Total Votes'].sum().sort_values(ascending=False).head(10).values)
plt.xticks(rotation=90)
plt.xlabel('Party')
plt.ylabel('Total Votes')
plt.title('Top 10 Political Parties by Total Votes')
plt.show()
```

```python
#Vote distribution by vote type
votes = [df['EVM Votes'].sum(), df['Postal Votes'].sum()]
labels = ['EVM Votes', 'Postal Votes']

print("EVM Votes:", df['EVM Votes'].sum())
print("Postal Votes:", df['Postal Votes'].sum())
```

```python
#This pie chart shows the distribution of votes cast through EVM and Postal Ballots. It helps to understand the contribution of each voting method to the total votes recorded in the election.
plt.figure(figsize=(10,6))
plt.pie(votes, labels=labels, autopct='%1.1f%%')
plt.title('Vote Distribution by Vote Type')
plt.show()
```

```python
#Top 10 constituencies by total votes
#This bar chart display the top 10 constituencies with the highest number of total votes cast in the election. It helps to identify constituencies with the highest voter participation and provides insights into voting trends.

top_10_constituencies = df.groupby('AC Name')['Total Votes'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.xticks(rotation=90)
plt.bar(top_10_constituencies.index, top_10_constituencies.values)
plt.xlabel('AC Name')
plt.ylabel('Total Votes')
plt.title('Top 10 Constituencies by Total Votes')
plt.show()
```

```python
#Vote share distribution
#This histogram shows the distribution of vote share percentages among candidates in the election. It helpds to analyze how votes were spread across candidates and identifies whether most candidates received similar vote shares or if a few candidates dominated the election with higher percentages.
# Vote share = Percentage of votes received by a candidate.
# Frequency = Number of candidates having that vote share range.

plt.figure(figsize=(10,6))
plt.hist(df['Vote Share'], bins=20)
plt.xlabel('Vote Share')
plt.ylabel('Frequency')
plt.title('Vote Share Distribution')
plt.show()
```

```python
#EVM votes vs postal votes
#This scatter plot shows the relationship between EVM votes and postal votes received by candidates in the election. Each point on the graph represents a candidate, where the horizontal axis shows EVM votes and the vertical axis shows postal votes. The visualisation helps analyze whether candidates receiving higher EVM votes also tend to receive higher postal votes.
plt.figure(figsize=(10,6))
plt.scatter(df['EVM Votes'], df['Postal Votes'])
plt.xlabel('EVM Votes')
plt.ylabel('Postal Votes')
plt.title('EVM Votes vs Postal Votes')
plt.show()
```

```python
#Constituency-wise winner analysis
#This bar chart shows the number of constituencies won by each political party in the election. It helps to identify the parties that secured the highest number of seats and provides insights into the overall election performance of different parties.

# Find the winner for each constituency
constituency_winners = df.loc[df.groupby('AC Name')['Total Votes'].idxmax()]

# Count how many constituencies each party won
party_wins = constituency_winners['Party'].value_counts().head(10)

plt.figure(figsize=(10,6))
party_wins.plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel('Party')
plt.ylabel('Number of Constituencies Won')
plt.title('Top 10 Parties by Number of Constituencies Won')
plt.show()
```

```python
#Top 10 winning candidates
#This horizontal bar chart displays the top 10 winning candidates based on the total number of votes received in the election. It helps  to identify the candidates with the strongest voter support and compares their performance.
winning_candidates = df.sort_values(by='Total Votes', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(winning_candidates['Candidate'], winning_candidates['Total Votes'])
plt.xlabel('Total Votes')
plt.ylabel('Candidate')
plt.title('Top 10 candidates by votes')
plt.show()
```
